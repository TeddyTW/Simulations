```python
import numpy as np
import matplotlib.pyplot as plt
import math
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
class ChargedParticle:
    def __init__(self, mass, charge, postion, velocity):
        self.m = mass
        self.q = charge
        self.r = np.array(postion)
        self.v = np.array(velocity)
        
    def PE_field(self, other_particles, E_factor=1, C_factor=0.1):
#         external=np.array([2*self.r[0], 2*self.r[1], -4*self.r[2]])
        external= [0,0,0]
        coulomb=0
        for i in other_particles:
            coulomb += 1*np.array((self.r-i.r)/np.linalg.norm(self.r-i.r)**3)
        return(np.array(external)+np.array(coulomb))
        
    def v_r_step(self, dt, E, B):
        
        
        # METHOD MORE SIMILAR TO YOURS
        v_minus = (dt*self.q*E)/(2*self.m) + self.v
        c = (dt*self.q)/(2*self.m)
        
        ## solve for v_plus through system of linear equations
        a = np.array([[1, -c*B[2], c*B[1]],[c*B[2], 1, -c*B[0]], [-c*B[1], c*B[0], 1]])
        #print(a)
        b = c * np.cross(v_minus, B) + v_minus
        #print(b)
        v_plus =np.linalg.solve(a, b)
        
        v_new = v_plus + (dt*self.q*E)/(2*self.m)
        # END OF METHOD SIMILAR TO YOURS
        
        # below is the original method, if your comment out the above method and uncomment the method
        # below you will see they give and equivalent dynamic
        
#         v_new=[]
#         Omega = (self.q*B)/self.m
#         Sigma = (self.q*E)/self.m
#         A = np.array((Omega*dt)/2)
#         C = np.array(self.v + dt*(Sigma + np.cross(self.v, Omega/2)))
#         v_new=(C + A*np.dot(A, C) - np.cross(A, C))/(1+np.dot(A, A))
#         #print(A)
        r_new = self.r + v_new*dt
        self.v = np.array(v_new)
        self.r = np.array(r_new)
```

```python
## Initialise particles

particles=[]
num_particles=2 #3
q= 1
m = 1
#Sara edited: B = [0,0,0]
B=np.array([0, 0, 5]) #original: [0,0,5])
for i in range(0, num_particles):
    #particles.append(ChargedParticle(m, q, np.random.normal(0, 0.1, 3), np.random.normal(0, 0.01, 3))) 
    # Sara edit: v=[0.001,0,0] r =[1,0,0]
    if i == 0:
        particles.append(ChargedParticle(m, q, [1, 0, 0], [0.001,0,0])) # original: [1,1,1], [0.001,0.001, 0.001]))
    if i == 1:
        particles.append(ChargedParticle(m, q, [-1, 0, 0], [0.001,0,0]))
    print(particles)
particles=np.array(particles)
```

```python
#iterate through steps0
T=0
V=[]
R=[]
dt=.01
time=[]

#print(x_old[1][0])
while T<10:
    time.append(T)
    r=[]
    v=[]
    for p in range(0, len(particles)):
        #print(p)
        #Sara edit: delete PE_field
        other_particles=np.delete(particles, p)
        E = particles[p].PE_field(other_particles)
        #E = np.array([0,0,0])
        #print('E=',E)
        particles[p].v_r_step(dt, E, B)
        #Sara edit: delete photon abs
        #particles[p].photon_absorption(3.8461538E+14)
        r.append(particles[p].r)
        v.append(particles[p].v)
    
    R.append(r)
    V.append(v)
    T=T+dt
```
